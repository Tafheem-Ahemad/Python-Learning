```python
# Assignment 8

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

transpose_matrix = [row.copy() for row in matrix]

for i in range(len(transpose_matrix)):
	for j in range (i+1):
		transpose_matrix[i][j] , transpose_matrix[j][i] = transpose_matrix[j][i], transpose_matrix[i][j]

print(matrix)
print(transpose_matrix)
```

```python
# Assignment 15 List Intersection 
li = [1 , 2 , 3, 4 ,5]
l2 = [4 , 5 , 6 , 7 ,  8]

l3 = [x for x in li if x in l2]
print(l3)
```
