```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
# Plot scater 

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('tips')
plt.figure(figsize=(5,5))
```

```python
# line plot

sns.lineplot(x='size',y='tip',data=tips)
plt.title('Sizes')
plt.figure(figsize=(5,5))
```

```python
# Catalogical plot

sns.barplot(x='sex',y='size',data=tips)
plt.title('Sizes')
plt.figure(figsize=(5,5))
```

```python
## Box Plot
sns.boxplot(y="day",x='total_bill',data=tips)
```

```python
## Violin Plot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
tips.head()
```

```python
# histograms
sns.histplot(x='total_bill',kde=True,bins=10,data=tips)
```

```python
# KDE
sns.kdeplot(x='total_bill',data=tips,fill=True)
```

```python

# Pairplot
sns.pairplot(tips)
```

```python
## HEatmap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(data=corr,annot=True,cmap='coolwarm')
```
