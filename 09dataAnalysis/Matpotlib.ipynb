```python
import matplotlib.pyplot as plt
```

```python
# line
x1=[1,2,3,4]
y1=[10,20,30,40]

plt.figure(figsize=(9,5)) # fixed dimentions
plt.plot(x1,y1,color="red",marker="o",linestyle="--")  #both x1,y1 has same size
plt.grid(True)
```

```python

# Multiplot
plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x1,y1,color="red")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("A")


plt.subplot(2,2,2)
plt.plot(x1,y1,color="red")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("A")

plt.subplot(2,2,3)
plt.plot(x1,y1,color="red")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("A")

plt.subplot(2,2,4)
plt.plot(x1,y1,color="red")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("A")

```

```python
x=['A','B','C']
y=[1,2,3]

plt.figure(figsize=(3,2))

plt.bar(x,y,color="red")
```

```python
d1={'A':10,'B':20,'C':30}
plt.figure(figsize=(3,2))

plt.bar(d1.keys(),d1.values(),color="red")
```

```python
# Histograms
li=[1,1,2,3,4,5,6,7,2,5,8]

plt.hist(li,bins=5,color='red',edgecolor='black')
```

```python
# PieChart
data=[10,40,25,4.8,20.2]

plt.pie(data,labels=['A','B','C','D','E'],explode=[.2,0,0,0,0],autopct="%1.0f%%")
# Explodes for some outing shape
```
